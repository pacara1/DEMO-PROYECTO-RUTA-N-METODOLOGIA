```python
import openpyxl
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
excel_document = openpyxl.load_workbook('files/MUERTOS.xlsx')
sheets=excel_document.get_sheet_names()
df5=pd.read_pickle('files/df5.pickle')
df5
```

# GRAFICOS CLASE DE ACCIDENTE ANUAL

```python
df5.dtypes
```

```python
data=df5.groupby([pd.Grouper(freq='Y', key='Fecha Ocurrencia'), 'Clase de Accidente']).sum().reset_index()
data['Fecha Ocurrencia']=data['Fecha Ocurrencia'].dt.year
data=data.rename(columns={'Fecha Ocurrencia':'Año Ocurrencia'})
data
```

```python
plt.figure(figsize=(15,60))
for index in range(len(sheets)):
    plt.subplot(len(sheets)/2, 2, index+1)
    y=data.loc[data.groupby('Año Ocurrencia').groups[list(data.groupby(data['Año Ocurrencia']).groups.keys())[index]]]
    plt.bar(range(len(y)), y['Muertes'])
    plt.xticks(range(len(y)),list(y.groupby('Clase de Accidente').groups.keys()),rotation='vertical')
    plt.title("Muertes por Clase de accidente/Anual"+"-"+sheets[index]);
plt.savefig("Muertes_Clase_Accidente_Anual.png")
```
