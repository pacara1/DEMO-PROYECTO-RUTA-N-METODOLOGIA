```python
df6=df5.dropna(subset=['Dirección Ocurrencia'])
df6.index=list(range(len(df6['Dirección Ocurrencia'])))
df6
```

```python
data=pd.DataFrame({'Año Ocurrencia':df6['Fecha Ocurrencia'].dt.year,
                 'Dirección Ocurrencia':df6['Dirección Ocurrencia'],
                 'Latitud':df6['Dirección Ocurrencia'].map(lambda x: str(x)) +', Medellin, Antioquia, Colombia',
                 'Longitud':df6['Dirección Ocurrencia'].map(lambda x: str(x)) +', Medellin, Antioquia, Colombia',
                 'Muertes':df6['Muertes']})

data
```

```python
import geocoder
for i in range (len(data['Dirección Ocurrencia'])):
    data['Latitud'][i]= geocoder.arcgis(data['Latitud'][i]).lat
    data['Longitud'][i]= geocoder.arcgis(data['Longitud'][i]).lng
data
```
